```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

# load and investigate the data here

```python
df = pd.read_csv("tennis_stats.csv")
df.head()
```

```python
df.columns
```

```python
df.describe()
```

# perform exploratory analysis here

```python
df.corr()
```

```python
plt.scatter(df['FirstServeReturnPointsWon'],df['Winnings'])
plt.title('FirstServeReturnPointsWon vs Winnings')
plt.xlabel('FirstServeReturnPointsWon')
plt.ylabel('Winnings')
plt.show()
plt.clf()
```

```python
plt.scatter(df['BreakPointsOpportunities'],df['Winnings'])
plt.title('BreakPointsOpportunities vs Winnings')
plt.xlabel('BreakPointsOpportunities')
plt.ylabel('Winnings')
plt.show()
plt.clf()
```

```python
plt.scatter(df['ServiceGamesPlayed'],df['Winnings'])
plt.title('ServiceGamesPlayed vs Winnings')
plt.xlabel('ServiceGamesPlayed')
plt.ylabel('Winnings')
plt.show()
plt.clf()
```

```python
plt.scatter(df['TotalPointsWon'],df['Winnings'])
plt.title('TotalPointsWon vs Winnings')
plt.xlabel('TotalPointsWon')
plt.ylabel('Winnings')
plt.show()
plt.clf()
```

## perform single feature linear regressions here

```python
# select BreakPointsOpportunities as a feature and winning as the outcome
x = df[['BreakPointsOpportunities']]
y = df[['Winnings']]

X_train, X_test, y_train, y_test = train_test_split(x, y, train_size = 0.8)

model = LinearRegression()
model.fit(X_train, y_train)

# score model on test data
print('Predicting Winnings with BreakPointsOpportunities Test Score:', model.score(X_test,y_test))


```

```python
y_pred = model.predict(X_test)
plt.scatter(y_test,y_pred, alpha=0.4)
plt.title('Predicted Winnings vs. Actual Winnings - BreakPointsOpportunities')
plt.xlabel('Actual Winnings')
plt.ylabel('Predicted Winnings')
plt.show()
plt.clf()
```

```python
# select ServiceGamesPlayed as a feature and winning as the outcome
x = df[['ServiceGamesPlayed']]
y = df[['Winnings']]

X_train, X_test, y_train, y_test = train_test_split(x, y, train_size = 0.8)

model = LinearRegression()
model.fit(X_train, y_train)

# score model on test data
print('Predicting Winnings with ServiceGamesPlayed Test Score:', model.score(X_test,y_test))
```

```python
y_pred = model.predict(X_test)
plt.scatter(y_test,y_pred, alpha=0.4)
plt.title('Predicted Winnings vs. Actual Winnings - ServiceGamesPlayed')
plt.xlabel('Actual Winnings')
plt.ylabel('Predicted Winnings')
plt.show()
plt.clf()
```

## perform two feature linear regressions here

```python
# select ServiceGamesPlayed and BreakPointsOpportunities as a feature and winning as the outcome
x = df[['BreakPointsOpportunities','ServiceGamesPlayed']]
y = df[['Winnings']]

X_train, X_test, y_train, y_test = train_test_split(x, y, train_size = 0.8)

model = LinearRegression()
model.fit(X_train, y_train)

# score model on test data
print('Predicting Winnings with ServiceGamesPlayed and BreakPointsOpportunities Test Score:', model.score(X_test,y_test))
```

```python
y_pred = model.predict(X_test)
plt.scatter(y_test,y_pred, alpha=0.4)
plt.title('Predicted Winnings vs. Actual Winnings - ServiceGamesPlayed and BreakPointsOpportunities')
plt.xlabel('Actual Winnings')
plt.ylabel('Predicted Winnings')
plt.show()
plt.clf()
```

## perform multiple feature linear regressions here

```python
x = df[['FirstServe','FirstServePointsWon','FirstServeReturnPointsWon',
'SecondServePointsWon','SecondServeReturnPointsWon','Aces',
'BreakPointsConverted','BreakPointsFaced','BreakPointsOpportunities',
'BreakPointsSaved','DoubleFaults','ReturnGamesPlayed','ReturnGamesWon',
'ReturnPointsWon','ServiceGamesPlayed','ServiceGamesWon','TotalPointsWon',
'TotalServicePointsWon']]

y = df[['Winnings']]

X_train, X_test, y_train, y_test = train_test_split(x, y, train_size = 0.8)

model = LinearRegression()
model.fit(X_train, y_train)

# score model on test data
print('Predicting Winnings with multiple features Test Score:', model.score(X_test,y_test))
```

```python
y_pred = model.predict(X_test)
plt.scatter(y_test,y_pred, alpha=0.4)
plt.title('Predicted Winnings vs. Actual Winnings - multiple features')
plt.xlabel('Actual Winnings')
plt.ylabel('Predicted Winnings')
plt.show()
plt.clf()
```
